# `ssh` remote execution: launch simulation task

```python
import paramiko

ssh = paramiko.SSHClient()
ssh.set_missing_host_key_policy(paramiko.AutoAddPolicy())
ssh.load_system_host_keys()
ssh.connect(hostname=host, username=user)
```

```python
def parse_io(io):
    io_ls = io.readlines()
    return "".join(io_ls)
```

```python
stdin, stdout, stderr = ssh.exec_command('pwd')  # Non-blocking call
exit_status = stdout.channel.recv_exit_status()          # Blocking call
if exit_status == 0:
    print ("completed")
else:
    print("Error", exit_status)
```

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command('cd D:\knpob; pwd')
print(parse_io(ssh_stdout))
```

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command('cd D:\knpob; ls')
print(parse_io(ssh_stdout))
```

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command('cd D:\knpob\e2x1; ls')
print(parse_io(ssh_stdout))
```

## Retrieve performance metrics

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command("(Get-Counter '\Processor(_Total)\% Processor Time').CounterSamples[0].CookedValue")
cpu_usage = float(parse_io(ssh_stdout))

ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command("(Get-WmiObject Win32_Processor).NumberOfCores")
cpu_cores = int(parse_io(ssh_stdout))
```

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command("(Get-CimInstance Win32_ComputerSystem).TotalPhysicalMemory / 1GB")
ram_total = float(parse_io(ssh_stdout))

ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command("(Get-Counter '\Memory\Available MBytes').CounterSamples[0].CookedValue")
ram_usage = ram_total - float(parse_io(ssh_stdout)) / 1024
```

```python
ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command(r"(Get-Counter '\Network Interface(*)\Bytes Total/sec').CounterSamples[0].CookedValue")
net_total = float(parse_io(ssh_stdout)) / 1024

ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command(r"(Get-Counter '\Network Interface(*)\Bytes Received/sec').CounterSamples[0].CookedValue")
net_rec = float(parse_io(ssh_stdout)) / 1024

ssh_stdin, ssh_stdout, ssh_stderr = ssh.exec_command(r"(Get-Counter '\Network Interface(*)\Bytes Sent/sec').CounterSamples[0].CookedValue")
net_send = float(parse_io(ssh_stdout)) / 1024
```

```python
import pandas as pd

data = {'Metric': ['CPU', 'RAM', 'Network'],
        'Usage': ['{:.2f}%'.format(cpu_usage), '{:.2f}%'.format(ram_usage / ram_total * 100), '{:.2f} KB/s'.format(net_total)],
        'Details': ['core number {}'.format(cpu_cores), 'used {:.2f} GB / total {:.2f} GB'.format(ram_usage, ram_total), 'send {:.2f} KB/s / receive {:.2f} KB/s'.format(net_send, net_rec)]}

pd.DataFrame(data)
```
